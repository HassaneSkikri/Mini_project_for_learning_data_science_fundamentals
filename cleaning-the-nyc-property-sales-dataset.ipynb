```python
import pandas as pd
import numpy as np

pd.set_option('display.max_column',None)
```

```python
# importing the data 
df = pd.read_csv('/kaggle/input/nyc-property-sales/nyc-rolling-sales.csv')
```

    ðŸ”»Step 2: Remove Unnecessary columnsðŸ”»
</p>


```python
# the column 'Unnamed: 0' is an iterator and is of no use here
df=df.drop(columns=['Unnamed: 0'])
#this is the same as df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.info()
```

```python
df.info()
```

<p style="background-color: #12f7ff; font-family: 'Trebuchet MS', sans-serif; color: #000; font-size: 150%; text-align: center; border-radius: 50px 15px; padding: 10px; box-shadow: 0 4px 8px 0 rgba(0, 0, 0, 0.2), 0 6px 20px 0 rgba(0, 0, 0, 0.19);">
    ðŸ”»Step 3: Whitespace and '-' handlingðŸ”»
</p>


```python
# the column 'EASE-MENT' is blank and can be dropped

df.drop('EASE-MENT', axis=1, inplace=True)
```

```python
df.head()
```

```python
#we need to discover if there is any empy space or any '-' caracter
# we will count the number of the empty spaces and '-' caracter
counts_ = df.applymap(lambda x: str(x).count("-"))
counts_space = df.applymap(lambda x: str(x).count(" "))
print(f"there is {counts_.sum().sum()} '-' and {counts_space.sum().sum()} spaces")
```

<p style="background-color: #12f7ff; font-family: 'Trebuchet MS', sans-serif; color: #000; font-size: 150%; text-align: center; border-radius: 50px 15px; padding: 10px; box-shadow: 0 4px 8px 0 rgba(0, 0, 0, 0.2), 0 6px 20px 0 rgba(0, 0, 0, 0.19);">
    ðŸ”»Step 4: Remove the duplicated valuesðŸ”»
</p>


```python
# if you see our data, it include somme missing values it represent it like '-' we will replace it by nan
# blank rows should instead be null values

map_dict={'-':np.nan,' ':np.nan}
df.replace(map_dict,inplace = True)
```

```python
#let's view if there is any duplicated values
sum(df.duplicated())
```

<p style="background-color: #12f7ff; font-family: 'Trebuchet MS', sans-serif; color: #000; font-size: 150%; text-align: center; border-radius: 50px 15px; padding: 10px; box-shadow: 0 4px 8px 0 rgba(0, 0, 0, 0.2), 0 6px 20px 0 rgba(0, 0, 0, 0.19);">
    ðŸ”»Step 6:Handle Non-Numeric Data in Numeric ColumnsðŸ”»
</p>


```python
numeric_columns = ['LAND SQUARE FEET', 'GROSS SQUARE FEET', 'SALE PRICE']
for col in df.columns:
    if col in numeric_columns:
        # Convert to numeric, setting errors to 'coerce' to handle non-numeric values
        df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.info()
```

<p style="background-color: #12f7ff; font-family: 'Trebuchet MS', sans-serif; color: #000; font-size: 150%; text-align: center; border-radius: 50px 15px; padding: 10px; box-shadow: 0 4px 8px 0 rgba(0, 0, 0, 0.2), 0 6px 20px 0 rgba(0, 0, 0, 0.19);">
    ðŸ”»Step 7: Impute Missing ValuesðŸ”»
</p>


```python
# Imputing missing values in 'LAND SQUARE FEET' and 'GROSS SQUARE FEET' using the median
for col in ['LAND SQUARE FEET', 'GROSS SQUARE FEET']:
    median_value = df[col].median()
    df[col].fillna(median_value, inplace=True)

# Removing rows where 'SALE PRICE' is missing
df = df.dropna(subset=['SALE PRICE'])

# Checking the changes made
df_info = {
    "Missing Values After Imputation": df[numeric_columns].isnull().sum(),
    "Remaining Rows": len(df)
}

df_info
```
